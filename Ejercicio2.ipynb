```python
from sklearn.datasets import fetch_openml
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")

iris = fetch_openml(name='iris', version=1, as_frame=True)

df = iris.frame
```

```python
# Mostrar los primeros registros del DataFrame
print("Primeros registros del dataset:")
print(df.head())

# Identificar el tipo de cada variable
print("\nTipos de variables en el dataset:")
print(df.dtypes)
```

```python
# Identificar si las variables son cuantitativas o cualitativas
print("\nClasificación de variables:")
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        print(f"{column}: Cuantitativa")
    else:
        print(f"{column}: Cualitativa")
```

```python
# Gráfico de Barras

plt.figure(figsize=(8, 6))
sns.countplot(x='class', data=df)
plt.title('Cantidad de Observaciones por Clase de Flor')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Gráfico de Sectores 

plt.figure(figsize=(8, 6))
df['class'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title('Proporción de Observaciones por Clase de Flor')
plt.ylabel('')  # Quitamos la etiqueta del eje y
plt.show()
```

```python
# Gráfico de Líneas

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['sepallength'], marker='o', linestyle='-', color='b')
plt.title('Sepal Length a lo Largo de las Observaciones')
plt.xlabel('Índice de Observación')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# Gráfico de Dispersión

plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepallength', y='petallength', hue='class', data=df, palette='viridis')
plt.title('Relación entre Sepal Length y Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(title='Clase de Flor')
plt.show()
```
